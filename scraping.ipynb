```python
import PyPDF2

print("PyPDF2 version:", PyPDF2.__version__)
```

```python
pdf_file_path = 'LBN_PenalCode1943_AR.pdf'

reader = PyPDF2.PdfReader(pdf_file_path)
print(f'The PDF has {len(reader.pages)} pages.')

# read first page
first_page = reader.pages[0]
text = first_page.extract_text()
print(text[:1000])
```

```python
# Extract all pages
all_text = ""

for page in reader.pages:
    page_text = page.extract_text()
    all_text += page_text + "\n"

print("✅ Total characters extracted:", len(all_text))

```

```python
# Look at a middle section of the text to inspect for noise
print(all_text[0:50500])
```

```python
import re
import unicodedata

# ranges for Arabic diacritics (Harakat)
HARAKAT_PATTERN = re.compile(r'[\u064B-\u0652\u0670\u0653-\u065F]')
TATWEEL_PATTERN = re.compile(r'\u0640')  # ـ

def clean_arabic_text_safe(text: str) -> str:
    # 1) Unicode normalization helps convert Arabic Presentation Forms to basic letters
    text = unicodedata.normalize("NFKC", text)

    # 2) Remove Tatweel and Harakat (optional but common in laws)
    text = TATWEEL_PATTERN.sub('', text)
    text = HARAKAT_PATTERN.sub('', text)

    # 3) Standardize newlines and spaces (keep paragraphs but avoid walls of \n)
    #    First, collapse multiple newlines to max 2 (paragraph feel)
    text = re.sub(r'\n{3,}', '\n\n', text)
    #    Replace runs of spaces/tabs with a single space
    text = re.sub(r'[ \t]+', ' ', text)
    #    Trim spaces around newlines
    text = re.sub(r' *\n *', '\n', text)

    # 4) Strip leading/trailing whitespace
    return text.strip()

cleaned_text = clean_arabic_text_safe(all_text)
print("✅ Total characters after cleaning:", len(cleaned_text))
print(cleaned_text[:800])
```

```python

def clean_extra_newlines(text):
    # Replace 3+ newlines with a single newline
    text = re.sub(r'\n{3,}', '\n\n', text)
    # Remove extra spaces before/after newlines
    text = re.sub(r' *\n *', '\n', text)
    # Strip leading/trailing spaces
    return text.strip()

# Apply on your existing cleaned_text
cleaned_text = clean_extra_newlines(cleaned_text)

print("✅ Cleaned extra newlines.")
print(cleaned_text[:500])
```

```python
save_path = r"C:\Users\PC\OneDrive\Desktop\chatbot\data\processed\penal_code_cleaned.txt"

with open(save_path, "w", encoding="utf-8") as f:
    f.write(cleaned_text)

print("✅ Saved cleaned penal code to:", save_path)
```

```python
# chunk the cleaned text into smaller pieces for easier processing
chunk_size = 800
overlap = 100

def chunk_text(text: str, chunk_size: int, overlap: int):
    chunks = []
    start = 0
    text_length = len(text)

    while start < text_length:
        end = min(start + chunk_size, text_length)
        chunk = text[start:end]
        chunks.append(chunk)

        # Move start forward by chunk_size - overlap
        start += chunk_size - overlap

    return chunks

chunks = chunk_text(cleaned_text, chunk_size, overlap)
print(f"✅ Total chunks created: {len(chunks)}")
print(chunks[0][:500])
```

```python
import json
import os

os.makedirs("../data/processed", exist_ok=True)

# Save all chunks into a JSON file
with open("../data/processed/penal_code_chunks.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, ensure_ascii=False, indent=2)

print("✅ Saved processed chunks to data/processed/penal_code_chunks.json")
```
